```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from math import radians, sin, cos, sqrt, atan2
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
import folium
from streamlit_folium import folium_static  # For map in app

# Upload CSV via Files panel if not done
df = pd.read_csv('/content/IIoT_Smart_Parking_Management.csv')  # Adjust filename
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
print(df.head())
print(df.info())
```

```python
df = df.drop_duplicates()
df = df.ffill()
df = df.sort_values(['Parking_Spot_ID', 'Timestamp'])

# Temporal features
df['Hour'] = df['Timestamp'].dt.hour
df['DayOfWeek'] = df['Timestamp'].dt.dayofweek
df['IsWeekend'] = (df['DayOfWeek'] >= 5).astype(int)
df['Month'] = df['Timestamp'].dt.month

# Spatial features
def haversine(row1, row2):
    # Assuming Lat and Lon columns exist or can be derived/are available
    # This function requires 'Lat' and 'Lon' which are not explicitly in the provided df info.
    # If these columns are missing, this part will fail.
    # For now, I'll keep it as is, assuming they might be added later or are implicitly handled.
    if 'Lat' not in row1 or 'Lon' not in row1 or 'Lat' not in row2 or 'Lon' not in row2:
        # This is a placeholder for handling missing Lat/Lon, actual implementation
        # depends on how Lat/Lon should be obtained.
        print("Warning: Lat or Lon columns missing for haversine calculation.")
        return 0.0 # Return a default or handle as appropriate
    lat1, lon1 = radians(row1['Lat']), radians(row1['Lon'])
    lat2, lon2 = radians(row2['Lat']), radians(row2['Lon'])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return 6371 * c

def add_distances(group):
    # This function also depends on 'Lat' and 'Lon'
    if 'Lat' not in group.columns or 'Lon' not in group.columns:
         print("Warning: Lat or Lon columns missing for distance calculation.")
         group['Dist_To_Nearest'] = 0.0
         return group

    group['Dist_To_Nearest'] = 0.0
    for idx, row in group.iterrows():
        distances = [haversine(row, group.iloc[i]) for i in range(len(group)) if group.index[i] != idx]
        group.at[idx, 'Dist_To_Nearest'] = min(distances) if distances else 0
    return group

# Applying add_distances requires 'Sensor_ID', 'Lat', and 'Lon'.
# 'Sensor_ID' is not in the provided df info, assuming 'Parking_Spot_ID' is intended.
# 'Lat' and 'Lon' are also not in the provided df info.
# The following line will likely cause an error if 'Parking_Spot_ID', 'Lat', or 'Lon' are not suitable or missing.
# I will replace 'Sensor_ID' with 'Parking_Spot_ID' based on available columns.
# Note: Spatial features calculation will only work if 'Lat' and 'Lon' columns are available or created.
if 'Parking_Spot_ID' in df.columns and 'Lat' in df.columns and 'Lon' in df.columns:
    df = df.groupby('Parking_Spot_ID').apply(add_distances).reset_index(drop=True)
else:
    print("Skipping spatial feature engineering due to missing 'Parking_Spot_ID', 'Lat', or 'Lon' columns.")
    df['Dist_To_Nearest'] = 0.0 # Add the column with default value to avoid errors later


# Encoding/Scaling
le = LabelEncoder()
if 'Vehicle_Type' in df.columns:
    df['Vehicle_Type_Encoded'] = le.fit_transform(df['Vehicle_Type'].fillna('None'))

scaler = MinMaxScaler()
# Updated num_cols based on available columns and potential new features
# Assuming 'Occupancy_Rate' is a numerical representation of occupancy.
num_cols = ['Hour', 'Occupancy_Rate', 'Dist_To_Nearest', 'Sensor_Reading_Proximity', 'Sensor_Reading_Pressure', 'Vehicle_Type_Weight', 'Vehicle_Type_Height', 'Weather_Temperature', 'Environmental_Noise_Level', 'Dynamic_Pricing_Factor', 'Proximity_To_Exit', 'User_Parking_History'] # Add other relevant numerical columns
available_num_cols = [col for col in num_cols if col in df.columns]
df[available_num_cols] = scaler.fit_transform(df[available_num_cols])

# Create sequences
def create_sequences(data, seq_len=24):
    X, y = [], []
    # Updated features to match available and scaled numerical columns
    features = ['Hour', 'Occupancy_Rate', 'Dist_To_Nearest', 'Sensor_Reading_Proximity', 'Sensor_Reading_Pressure', 'Vehicle_Type_Weight', 'Vehicle_Type_Height', 'Weather_Temperature', 'Environmental_Noise_Level', 'Dynamic_Pricing_Factor', 'Proximity_To_Exit', 'User_Parking_History'] # Use scaled features
    available_features = [f for f in features if f in data.columns]
    if 'Occupancy_Status' not in data.columns:
        print("Error: 'Occupancy_Status' column not found for target variable.")
        return np.array([]), np.array([]) # Return empty arrays if target is missing

    # Convert 'Occupancy_Status' to numerical if needed, assuming 'Occupied' is 1 and 'Vacant' is 0
    data['Occupancy_Status_Encoded'] = data['Occupancy_Status'].apply(lambda x: 1 if x == 'Occupied' else 0)

    for i in range(len(data) - seq_len):
        X.append(data.iloc[i:i+seq_len][available_features].values)
        y.append(data.iloc[i+seq_len]['Occupancy_Status_Encoded']) # Use encoded occupancy status as target
    return np.array(X), np.array(y)

X, y = create_sequences(df)

# Check if sequences were created before splitting
if X.size > 0 and y.size > 0:
    split_idx = int(0.8 * len(X))
    X_train, X_test = X[:split_idx], X[split_idx:]
    y_train, y_test = y[:split_idx], y[split_idx:]
    print(f"Data shapes: X_train={X_train.shape}, y_train={y_train.shape}, X_test={X_test.shape}, y_test={y_test.shape}")
else:
    print("Could not create sequences. Check data and column availability.")
    # Initialize with empty arrays to prevent errors in subsequent cells
    X_train, X_test, y_train, y_test = np.array([]), np.array([]), np.array([]), np.array([])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import joblib

# Random Forest
X_train_rf = X_train.reshape(X_train.shape[0], -1)
X_test_rf = X_test.reshape(X_test.shape[0], -1)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_rf, y_train)
y_pred_rf = rf_model.predict(X_test_rf)
print(f"RF Accuracy: {accuracy_score(y_test, y_pred_rf):.4f}")
print(f"RF F1: {f1_score(y_test, y_pred_rf):.4f}")

# LSTM
lstm_model = Sequential()
lstm_model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
lstm_model.add(Dropout(0.2))
lstm_model.add(LSTM(50))
lstm_model.add(Dense(1, activation='sigmoid'))
lstm_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
lstm_model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2, verbose=0)

# Save models
joblib.dump(rf_model, 'rf_model.pkl')
lstm_model.save('lstm_model.h5')
joblib.dump(scaler, 'scaler.pkl')
print("Models saved!")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
import joblib
import folium
from streamlit_folium import folium_static
import os
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Page config for professional wide layout
st.set_page_config(
    page_title="Smart Parking Predictor",
    page_icon="https://elements.lbl.gov/wp-content/uploads/2022/08/220826_Noun-Parking-Dark_703x469.png",
    layout="wide",
    initial_sidebar_state="expanded"
)

# Custom theme for professional look (blue-green urban theme)
st.markdown("""
    <style>
    .main .block-container {padding-top: 2rem;}
    .stMetric {background-color: #f0f2f6; padding: 1rem; border-radius: 0.5rem;}
    .stSelectbox > label {font-weight: bold; color: #1f77b4;}
    </style>
""", unsafe_allow_html=True)

# Load models
@st.cache_resource
def load_models():
    lstm_model = load_model('lstm_model.h5')
    scaler = joblib.load('scaler.pkl')
    return lstm_model, scaler

lstm_model, scaler = load_models()

# Initialize session state for prediction results
if 'status' not in st.session_state:
    st.session_state.status = None
if 'pred' not in st.session_state:
    st.session_state.pred = None
if 'prediction_made' not in st.session_state:
    st.session_state.prediction_made = False

# Header
st.header("Smart Parking Slot Predictor Dashboard")
st.markdown("---")

# Sidebar for inputs (professional grouping)
with st.sidebar:
    st.subheader("üìä Input Parameters")
    vehicle_type = st.selectbox("Vehicle Type", ['Car', 'SUV', 'Truck'], help="Select your vehicle for size recommendation")
    size_map = {'Car': 'Small Slot (2.5m x 5m)', 'SUV': 'Medium Slot (3m x 5.5m)', 'Truck': 'Large Slot (3.5m x 6m)'}
    st.info(f"**Recommended Slot Size**: {size_map[vehicle_type]}")

    slot_list = [f"SLOT_{i:03d}" for i in range(1, 51)]  # SLOT_001 to SLOT_020
    slot_id = st.selectbox("Slot ID", slot_list, help="Choose a parking slot")

    hour = st.slider("Hour (0-23)", 0, 23, 12, help="Time of day for prediction")
    sensor_val = st.slider("Sensor Value (0-1)", 0.0, 1.0, 0.5, help="IoT sensor reading (higher = more likely occupied)")

    # Email setup (optional, for notification)
    sender_email = st.text_input("Sender Email (Gmail)", "your_email@gmail.com", help="For notifications")
    app_password = st.text_input("App Password", type="password", help="Gmail app password (enable 2FA)")
    receiver_email = st.text_input("Your Email for Notification")

# Main area: Traffic and Prediction (columns for balance)
col1, col2 = st.columns(2)

with col1:
    # Enhanced Traffic Indication (progress bar)
    traffic_level = "High" if (7 <= hour <= 9 or 17 <= hour <= 19) else "Low"
    traffic_color = "inverse" if traffic_level == "High" else "normal"
    st.metric(
        label="Traffic Level",
        value=traffic_level,
        delta="Peak hours detected" if traffic_level == "High" else "Off-peak",
        delta_color="warning" if traffic_level == "High" else "normal"
    )

with col2:
  if st.button("üîÆ Predict Availability", type="primary", use_container_width=True):
    # ---- Corrected Input Features (12 total) ----
    # Matching the scaler and LSTM training setup
    # Features: ['Hour', 'Occupancy_Rate', 'Dist_To_Nearest', 'Sensor_Reading_Proximity',
    #            'Sensor_Reading_Pressure', 'Vehicle_Type_Weight', 'Vehicle_Type_Height',
    #            'Weather_Temperature', 'Environmental_Noise_Level', 'Dynamic_Pricing_Factor',
    #            'Proximity_To_Exit', 'User_Parking_History']

    # Dummy or default representative values for prediction
    # You can later replace these with real-time inputs or sensor data
    feature_vector = [
        hour,                   # Hour (user input)
        0.5,                    # Occupancy_Rate (assumed)
        0.1,                    # Dist_To_Nearest
        sensor_val,             # Sensor_Reading_Proximity (user input)
        0.4,                    # Sensor_Reading_Pressure
        0.5,                    # Vehicle_Type_Weight
        0.6,                    # Vehicle_Type_Height
        30,                     # Weather_Temperature (¬∞C)
        40,                     # Environmental_Noise_Level (dB)
        0.7,                    # Dynamic_Pricing_Factor
        0.3,                    # Proximity_To_Exit
        0.2                     # User_Parking_History
    ]

    # Create 24 time steps of the same feature vector (sequence)
    input_seq = np.array([feature_vector] * 24)  # shape = (24, 12)

    # Scale all 12 features with the trained MinMaxScaler
    input_seq_scaled = scaler.transform(input_seq).reshape(1, 24, 12)

    # Make LSTM prediction
    pred = lstm_model.predict(input_seq_scaled, verbose=0)[0][0]
    status = "Free" if pred < 0.5 else "Occupied"

    # Store results in session state
    st.session_state.status = status
    st.session_state.pred = pred
    st.session_state.prediction_made = True

    # Display results
    st.success(f"**Prediction for {slot_id} ({vehicle_type})**: {status}")
    st.metric("Availability Probability", f"{pred:.2%}",
              delta=f"{'Likely Free' if pred < 0.5 else 'Likely Occupied'}")




# Fixed: Use session state for reservation (outside predict block)
if st.session_state.prediction_made and st.session_state.status == "Free":
    if st.button("üìß Reserve & Notify", type="secondary", use_container_width=True):
        if sender_email and app_password and receiver_email:
            try:
                msg = MIMEMultipart()
                msg['From'] = sender_email
                msg['To'] = receiver_email
                msg['Subject'] = f'Parking Reservation for {slot_id}'
                body = f"Slot {slot_id} reserved at {hour}:00! Status: {st.session_state.status}. Vehicle: {vehicle_type}. Probability: {st.session_state.pred:.2f}."
                msg.attach(MIMEText(body, 'plain'))
                server = smtplib.SMTP('smtp.gmail.com', 587)
                server.starttls()
                server.login(sender_email, app_password)
                text = msg.as_string()
                server.sendmail(sender_email, receiver_email, text)
                server.quit()
                st.success("Reservation confirmed and email sent!")
            except Exception as e:
                st.error(f"Email failed: {str(e)}. Check credentials.")
        else:
            st.warning("Enter email details to send notification.")
else:
    if st.session_state.prediction_made:
        st.warning("Slot not available for reservation‚Äîtry another prediction.")

# Expandable results section
with st.expander("üìà Detailed Results", expanded=False):
    if st.session_state.prediction_made:
        st.write(f"**Vehicle Fit**: {size_map[vehicle_type]} ‚Äì Optimized for {vehicle_type} dimensions.")
        st.write(f"**Traffic Insight**: {traffic_level} traffic at hour {hour} ‚Äì Expect {'delays' if traffic_level == 'High' else 'smooth access'}.")

# New: Google Map for parking slots (with error handling)
@st.cache_data
def load_and_create_map():
    csv_path = 'IoT_Smart_Parking_Management.csv'
    if not os.path.exists(csv_path):
        st.warning(f"CSV file '{csv_path}' not found. Upload it to /content/ and refresh. Using dummy data for demo.")
        dummy_df = pd.DataFrame({
            'Lat': [40.7128, 40.7130, 40.7125],
            'Lon': [-74.0060, -74.0058, -74.0062],
            'Occupancy': [0, 1, 0],
            'Slot_ID': ['SLOT_001', 'SLOT_002', 'SLOT_003']
        })
        df_map = dummy_df
    else:
        df_map = pd.read_csv(csv_path)
        df_map['Timestamp'] = pd.to_datetime(df_map['Timestamp'])

        # Robust column handling for 'Occupancy'
        occupancy_col = None
        possible_cols = ['Occupancy', 'Status', 'Occupied', 'Availability']
        for col in possible_cols:
            if col in df_map.columns:
                occupancy_col = col
                break
        if occupancy_col:
            df_map[occupancy_col] = pd.to_numeric(df_map[occupancy_col], errors='coerce')
            df_map = df_map.rename(columns={occupancy_col: 'Occupancy'})
        else:
            st.warning("No 'Occupancy' or similar column found. Using dummy data for map.")
            dummy_df = pd.DataFrame({
                'Lat': [40.7128, 40.7130, 40.7125],
                'Lon': [-74.0060, -74.0058, -74.0062],
                'Occupancy': [0, 1, 0],
                'Slot_ID': ['SLOT_001', 'SLOT_002', 'SLOT_003']
            })
            df_map = dummy_df

        df_map = df_map.fillna(0)

    # Create interactive map
    m = folium.Map(location=[df_map['Lat'].mean(), df_map['Lon'].mean()], zoom_start=15, tiles='OpenStreetMap')
    for idx, row in df_map.head(10).iterrows():
        color = 'green' if row['Occupancy'] == 0 else 'red'
        folium.CircleMarker(
            location=[row['Lat'], row['Lon']],
            radius=8,
            popup=f"Slot {row['Slot_ID']}: {'Free' if row['Occupancy'] == 0 else 'Occupied'}",
            color=color,
            fill=True,
            fillColor=color,
            fillOpacity=0.7,
            tooltip=f"Click for details: {row['Slot_ID']}"
        ).add_to(m)
    return m

st.subheader("üó∫Ô∏è Interactive Parking Lot Map")
col_map1, col_map2 = st.columns(2)
with col_map1:
    folium_map = load_and_create_map()
    folium_static(folium_map, width=600, height=400)
with col_map2:
    st.write("**Map Legend**")
    st.markdown("""
    - üü¢ **Green Marker**: Free Slot
    - üî¥ **Red Marker**: Occupied Slot
    - Click markers for popups with slot details.
    """)
    st.info("Upload the CSV for real-time slot data; demo uses sample locations.")

# Footer with summary
st.markdown("---")
col_footer1, col_footer2 = st.columns(2)
with col_footer1:
    st.caption("Powered by LSTM & Random Forest Models")
with col_footer2:
    st.caption("Dataset: Smart Parking Management IoT (2021-2024)")
```
